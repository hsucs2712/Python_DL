```python
# 旋轉影像處理
from skimage.transform import rotate
import matplotlib.pyplot as plt
malaria = plt.imread('00000.jpg')
plt.figure(figsize=(2,2))
plt.imshow(malaria)
plt.axis('off')
plt.show()
```

```python
for a in range(0, -370, -10):
    print('Angle', a)
    plt.figure(figsize=(2,2))
    plt.imshow(rotate(malaria, angle=a))
    plt.axis('off')
    plt.show()
```

```python
#等比例調整影像大小
from skimage.transform import rescale
# 	rescale(image, 1/4, anti_aliasing=True, multichannel=True)
import matplotlib.pyplot as plt
andy = plt.imread('andy2.jpg')
plt.imshow(andy)
plt.show()

downscale = 1/2
andy_rescale = rescale(andy, downscale, anti_aliasing=True, multichannel=True)
plt.imshow(andy_rescale)
plt.show()

andy_rescale_anti = rescale(andy, downscale, anti_aliasing=False, multichannel=True)
plt.imshow(andy_rescale_anti)
plt.show()
```

```python
# 任意比例調整影像大小
from skimage.transform import resize
andy_resize = resize(andy, (400, 800), anti_aliasing=True)
plt.imshow(andy_resize)
plt.show()
```
