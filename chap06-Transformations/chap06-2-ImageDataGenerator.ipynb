```python
import io
import numpy as np
from PIL import Image
from keras.preprocessing.image import ImageDataGenerator
src = Image.open('00000.jpg')
img = np.array(src)
src = img.reshape((1,) + img.shape) 
datagen = ImageDataGenerator(  rotation_range=180,       #相片選轉角度
                               samplewise_center=True,                               
                               width_shift_range=0.01,
                               height_shift_range=0.01,
                               shear_range=0.03,     
                               zoom_range=[0.85, 1.25],     # 放大縮小
                               horizontal_flip=True,        # 影像水平反轉
                               vertical_flip=False,         # 影像垂直反轉
                               fill_mode="nearest",                               
                               brightness_range=[0.5, 1.5], #白皙亮光
                               rescale=1./255
                            )
i = 0
for batch in datagen.flow(src, batch_size=1, save_to_dir="./images", 
                          save_prefix="00000", save_format="jpg"):
    i += 1
    if i >= 5:
        break 
```
