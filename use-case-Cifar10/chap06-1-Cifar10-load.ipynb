```python
#Import libraries

import os
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit

#Download dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
```

```python
# 產生標籤
name2num = {'airplane' : 0,
            'automobile' : 1,
            'bird' : 2,
            'cat' : 3,
            'deer' : 4,
            'dog' : 5,
            'frog' : 6,
            'horse' : 7,
            'ship' : 8,
            'truck' : 9}
num2name = dict([(v,k) for (k,v) in name2num.items()])
num2name
```
