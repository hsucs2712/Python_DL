```python
import matplotlib.pyplot as plt
from skimage.util import random_noise

andy = plt.imread('andy.jpg')
plt.imshow(andy)
plt.show()
modes = ['gaussian', 'localvar', 'poisson', 'salt', 'pepper', 's&p', 'speckle' ]
for m in modes:
    andy_noise = random_noise(andy, mode= m)
    plt.title(m)
    plt.imshow(andy_noise)
    plt.show()
```

```python
andy_noisy = random_noise(andy, mode= modes[0])
plt.title(modes[0])
plt.imshow(andy_noisy)
plt.show()
```

```python
from skimage.restoration import denoise_tv_chambolle
import numpy as np
# 
for w in np.linspace(0.01, 0.5, 10):
    denoise_andy = denoise_tv_chambolle(andy_noisy,  weight=w, multichannel=True)
    plt.title('weight '+str(w))
    plt.imshow(denoise_andy)
    plt.show()
```

```python
########### denoise_bilateral ###############
from skimage.restoration import denoise_bilateral
import numpy as np
modes = ['constant', 'edge', 'symmetric', 'reflect', 'wrap']
for m in modes:
    denoise_andy = denoise_bilateral(andy_noisy, mode=m, multichannel=True)
    plt.title('Mode ' + m)
    plt.imshow(denoise_andy)
    plt.show()
```
