```python
from skimage.restoration import denoise_tv_chambolle
import numpy as np

for w in np.linspace(0.01, 0.5, 10):
    denoise_andy = denoise_tv_chambolle(andy,  weight=w, multichannel=True)
    plt.figure(figsize=(6,8))
    plt.title('weight '+str(w))
    plt.imshow(denoise_andy)
    plt.show()
```

```python
########### denoise_bilateral ###############
from skimage.restoration import denoise_bilateral
import numpy as np
modes = ['constant', 'edge', 'symmetric', 'reflect', 'wrap']
for m in modes:
    denoise_andy = denoise_bilateral(andy, mode=m, multichannel=True)
    plt.figure(figsize=(6,8))
    plt.title('Mode '+m)
    plt.imshow(denoise_andy)
    plt.show()
```
