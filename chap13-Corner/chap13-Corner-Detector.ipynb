```python
# --Corner detection 找出邊角 -- 轉換灰階
# --harris corner detector
from skimage.feature import corner_harris, corner_peaks
from skimage import color, io
import matplotlib.pyplot as plt
tp    = io.imread('building.jpg')
tp101 = io.imread('building.jpg', as_gray=True)
plt.imshow(tp101, cmap='gray')
plt.show()
```

```python
corner_harris?
```

```python
tp101_harris = corner_harris(tp101, method='eps')
plt.imshow(tp101_harris, cmap='gray_r')
plt.show()
```

```python
corner_peaks?
```

```python
corners = corner_peaks(tp101_harris, min_distance = 55,  threshold_rel=0.1)
print(len(corners))
corners[:3]
```

```python
def show_image_with_corners(image, coords, title="Corners detected"):
    plt.imshow(image, interpolation='nearest', cmap='gray')
    plt.title(title)
    plt.plot(coords[:, 1], coords[:, 0], '+r', markersize=15)
    plt.axis('off')
    plt.show()
```

```python
show_image_with_corners(tp, corners)
```
