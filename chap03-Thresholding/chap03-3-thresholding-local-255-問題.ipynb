```python
# 講解
import matplotlib.pyplot as plt
from skimage import color
andy = color.rgb2gray(plt.imread('andy.jpg')) #* 255.0 #<----------------很重要 因為彩色轉灰階會將數值0-1需要放大

#andy = plt.imread('andy.jpg')

plt.imshow(andy, cmap='gray')
plt.show()
thresh = 100/255.0
thresh_image = andy > thresh

plt.imshow(thresh_image, cmap='gray')
plt.show()
print(andy.min(), andy.max())
```

```python
from skimage.filters import try_all_threshold
fg, ax = try_all_threshold(andy, verbose = False)
```

```python
from skimage.filters import threshold_local
# andy = andy * 255.0
local_value = threshold_local(image=andy, block_size=35, offset=9) 
print(local_value.shape)
local_andy = andy > local_value
plt.imshow(local_andy, cmap='gray')
plt.show()
```
