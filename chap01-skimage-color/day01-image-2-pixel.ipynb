```python
import matplotlib.pyplot as plt
# dir(plt)
bear = plt.imread('./bear.jpg')
print(bear.shape, bear.dtype)
plt.imshow(bear)
plt.show()
```

```python
bear.flatten()
```

```python
bear.min(), bear.max()
```

```python
plt.hist(bear.flatten(), edgecolor='red' )
plt.show()
```
