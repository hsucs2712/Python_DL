```python
import matplotlib.pyplot as plt
# dir(plt)
bird2 = plt.imread('./bird.png')
print(bird2.shape, bird2.dtype)
plt.imshow(bird2)
plt.show()
```

```python
# RGB 2 Gray
from skimage import color
bird_gray = color.rgb2gray(bird2[:, :, :3])
bird_gray.min(), bird_gray.max()
plt.imshow(bird_gray, cmap='gray_r')
plt.show()
```
