```python
# --找邊界Edge -- 先轉為 Gray -- 才可以處理
# from skimage.filters import sobel
# --找邊界Edge(Sobel , Canny) -- 需要轉為灰階
# from skimage.feature import canny

%matplotlib inline
from skimage import io
img = io.imread('peoples.jpg', as_gray=True)
io.imshow(img)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from skimage.feature import canny
img_canny = canny(img, sigma=4)
io.imshow(img_canny.astype('float32'))
plt.show()
```

```python
%matplotlib inline
from skimage.filters import sobel
img_sobel = sobel(img)
io.imshow(img_sobel.astype('float32'))
plt.show()
```

```python
# contours
from skimage.filters import threshold_otsu
import matplotlib.pyplot as plt
thres = threshold_otsu(img)
print(thres)
img_thres = img >  thres
plt.imshow(img_thres, cmap='gray')
plt.axis('off')
plt.show()
```

```python
from skimage.measure import find_contours
contours = find_contours(img_thres, level=0.1 )
```

```python
def show_image_contour(image, contours):
    plt.figure()
    for n, contour in enumerate(contours):
        plt.plot(contour[:, 1], contour[:, 0], linewidth=2, color='red')
    plt.imshow(image, interpolation='nearest', cmap='gray')
    plt.title('Contours')
    plt.axis('off')
    plt.show()
```

```python
import matplotlib.pyplot as plt
show_image_contour(img_thres, contours) # 圖框的大小範圍
```
