```python
import matplotlib.pyplot as plt
colors = ['r', 'g', 'b']
img = plt.imread('fruits.jpg')
for rgb in [0, 1, 2]:
    plt.figure(figsize=(12, 3))
    plt.title('Red')
    plt.imshow(img[:, :, rgb], cmap='gray')
    plt.show()
    plt.figure(figsize=(12, 3))
    plt.hist( img[:, :, rgb].flatten(), color=colors[rgb])
    plt.show()
```
