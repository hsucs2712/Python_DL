```python
# RGBA是一種色彩空間的模型，由RGB色彩空間和Alpha通道組成。
# RGBA代表紅（Red）、綠（Green）、藍（Blue）和Alpha通道（Alpha）
import matplotlib.pyplot as plt
import numpy as np
dogs = plt.imread('two_dogs.png')
print(dogs.shape) # RGBA
plt.imshow(( dogs[:, :, :-1] ), cmap='gray') # <--RGBA 0123

plt.axis('off')
plt.show()
plt.imshow(np.fliplr( dogs[:, :, :-1] ), cmap='gray') # <--RGBA 0123
plt.axis('off')
plt.show()
```

```python
# RGB
import matplotlib.pyplot as plt
fruits = plt.imread('fruits.jpg')
plt.imshow(fruits)
plt.axis('off')
plt.show()
```

```python
plt.imshow(fruits[:, :, 0], cmap='gray')
plt.title('Red')
plt.show()
plt.imshow(fruits[:, :, 1], cmap='gray')
plt.title('Green')
plt.show()
plt.imshow(fruits[:, :, 2], cmap='gray')
plt.title('Blue')
plt.show()
```

```python
plt.hist(fruits[:, :, 0].flatten(), color='red')
plt.title('Red')
plt.show()
plt.hist(fruits[:, :, 1].flatten(), color='green')
plt.title('Green')
plt.show()
plt.hist(fruits[:, :, 2].flatten(), color='blue')
plt.title('Blue')
plt.show()
```
