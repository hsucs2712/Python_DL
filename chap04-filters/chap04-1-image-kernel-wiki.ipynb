```python
# https://en.wikipedia.org/wiki/Kernel_(image_processing)
# image kernel process = convolve2d
import matplotlib.pyplot as plt
from scipy import misc, signal 
import numpy as np
im = plt.imread('Vd-Orig.png')
```

```python
# sharpen_kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])             # Sharpen
# sharpen_kernel = np.array([ -1, -1, -1, -1, 8, -1, -1, -1 ,-1]).reshape(3,3) # Ridge detection
# sharpen_kernel = np.array([1, 2, 1, 2, 4, 2, 1, 2, 1]).reshape(3,3)/16       # Gaussian blur 
# sharpen_kernel = np.array([0, 1, 0, 1, -4, 1, 0, 1, 0]).reshape(3,3)
sharpen_kernel = np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]])                 #Identity
im_sharpened = np.ones(im.shape)
for i in range(3):
    im_sharpened[..., i] = np.clip(signal.convolve2d(im[...,i], 
                                                     sharpen_kernel, 
                                                     mode='same', 
                                                     boundary="fill"), 0, 1)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6, 6))
ax[0].imshow(im)
ax[0].set_title('Original Image', size=12)
ax[1].imshow(im_sharpened)
ax[1].set_title('Convolve2d Image', size=12)
plt.show()
```
