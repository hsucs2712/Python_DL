```python
data = [[1, 2, 3], [4], [5, 6], [7, 8, 9]]
# [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
# for loop
for dd in data:
    for d in dd:
        print(d)
# list comprehension
[d for dd in data for d in dd  ]
```

```python
data2 = [[1, 2, 3], 4, [5, 6], 7, 8, 9]
# [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
# 補充
sample = [1, 3, 5, 6, 8]

[s for s in sample if s%2==0] #假如僅有if
```

```python
# 補充
sample = [1, 3, 5, 6, 8]

[0 if s%2==0 else 1 for s in sample ] #假如有if else
```

```python
#多層
for ii in [[1, 2], [3, 4]]:
    for i in ii:
        print(i)

[i for ii in [[1, 2], [3, 4]] for i in ii ]        
```

```python
# for loop
for dd in data2:
    # print(dd)
    if type(dd)!=list:
        print(dd)
# list comprehension
###########################################################
data3 = [ [dd] if type(dd)!=list else dd for dd in data2  ]
[d for dd in data3 for d in dd  ]
###########################################################
[d for dd in [ [dd] if type(dd)!=list else dd for dd in data2  ] for d in dd  ]
```

```python
# 練習
stopword  = [',', '.', '!', '%']
statement = ['Hello World, I am % and see you then.', 'I am so great to see you!!!']
# 請用list comprehension 除去stopword
''.join([s for s in statement[1] if s not in stopword])
```
