```python
from skimage.filters import gaussian
from matplotlib import pyplot as plt
sigma = range(0, 10, 2)
for s in sigma:
    gaussian_sobel = gaussian(andy, sigma=s, multichannel=True)
    plt.imshow(gaussian_sobel)
    plt.title('Sigma = ' + str(s))
    plt.show()
```

```python
from skimage.filters import gaussian
from matplotlib import pyplot as plt
mode = ['reflect', 'constant', 'nearest', 'mirror', 'wrap']
for m in mode:
    gaussian_sobel = gaussian(andy, sigma=2, multichannel=True, mode=m)
    plt.title('Mode '+ m)
    plt.imshow(gaussian_sobel)
    plt.show()
```
