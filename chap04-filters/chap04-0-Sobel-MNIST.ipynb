```python
# from keras.models import Sequential
from keras.datasets import mnist

# 載入 MNIST 資料庫的訓練資料，並自動分為『訓練組』及『測試組』
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[0].reshape(28, 28), cmap='gray')
plt.show()
```

```python
from skimage.filters import sobel
mnist_sobel = sobel(image = X_train[0].reshape(28, 28))
plt.imshow(mnist_sobel, cmap='gray')
plt.show()
```

```python
# 不可以這樣做 實作錯誤導致整個所有影像數據mixed
from skimage.filters import sobel
mnist_sobel_all = sobel(image = X_train[:]) # <-這邊實作錯誤導致整個所有影像數據mixed
plt.imshow(mnist_sobel_all[0], cmap='gray')
plt.show()
mnist_sobel_all.shape
```

```python
####################################################
# 怎樣解決上述的實作錯誤 進行所有 影像數據的 sobel處理 
####################################################
from skimage.filters import sobel
import numpy as np
mnist_sobel_all = np.array([sobel(image = x) for x in X_train[:]])
plt.imshow(mnist_sobel_all[0], cmap='gray')
plt.show()
mnist_sobel_all.shape
```

```python
# 何謂 list comprehension - set operation
import numpy as np
data = [55, 44, 33, 77, 66]
np.array(data) ** 0.5 * 10 
```

```python
########################################
# 何謂 list comprehension - for loop with individual
########################################
import numpy as np
data = [55, 44, 33, 77, 66]
for d in data:

    print(d ** 0.5 * 10)
```

```python
########################################
# 何謂 list comprehension 
########################################
import numpy as np
data = [55, 44, 33, 77, 66]
[d ** 0.5 * 10 for d in data]
```

```python
# 人臉邊界
import matplotlib.pyplot as plt
from skimage.filters import sobel
andy_org = plt.imread('andy.jpg')
plt.imshow(andy_org)
plt.show()


thresh = 120/255.0
from skimage import io
andy = io.imread('andy.jpg', as_gray=True)

binary = (andy > thresh )

plt.imshow(binary, cmap= 'gray')
plt.show()

andy_sobel = sobel(image = binary) #<--------------這邊可以實作 andy, binary
plt.imshow(andy_sobel)
plt.show()

# help(sobel)
```
