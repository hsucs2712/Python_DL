```python
import matplotlib.pyplot as plt
sea = plt.imread('sea.jpg')
plt.imshow(sea)
plt.show()
```

```python
sea_copy = sea.copy()
sea_copy[200:600, 790:810, :] = 0
plt.imshow(sea_copy)
plt.show()
```

```python
import numpy as np
mask = np.zeros( (sea_copy.shape[0], sea_copy.shape[1]) )
mask[200:600, 790:810] = 1
plt.imshow(mask, cmap='gray')
plt.show()
```

```python
from skimage.restoration import inpaint
restored_image = inpaint.inpaint_biharmonic(sea_copy, mask, multichannel=True)
plt.imshow(restored_image)
plt.show()
```
