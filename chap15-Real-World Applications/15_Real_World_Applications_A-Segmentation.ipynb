```python
from skimage import segmentation
from skimage.color import label2rgb

segments = segmentation.slic(andy, start_label = 1, n_segments=100)

andy_segments = label2rgb(label=segments, image=andy, kind='avg', bg_label=None) # {'overlay', 'avg'}
print(andy_segments.min(), andy_segments.max())

andy_segments = andy_segments.astype('uint8') #

print(andy_segments.min(), andy_segments.max())

plt.imshow(andy_segments)
plt.show()
# Clipping input data to the valid range for imshow with RGB data ([0..1] for floats or [0..255] for integers).
```
