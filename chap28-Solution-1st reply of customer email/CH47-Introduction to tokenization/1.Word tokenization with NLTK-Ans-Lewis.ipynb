```python
######################################################################
from nltk.corpus import stopwords  
import string
import spacy
# 怎樣導入 stopwords ##################################################
stop_words_nltk = set(stopwords.words('english')) 
spacy.load('en_core_web_sm')
my_stop_word = set(['#', ',', '?', '1', '.', '!', 'the', 'of', ':', 'a', '2', 'I', "'s", 'It', '...'])
exclude = set(string.punctuation)
stop_words_spacy = set(spacy.lang.en.stop_words.STOP_WORDS) # spacy 停滯詞
stopword = (my_stop_word | stop_words_nltk | exclude | stop_words_spacy)
# 怎樣導入 stopwords ##################################################

import matplotlib.pyplot as plt
import pandas as pd
words = word_tokenize(scene_one.replace(':',''))
words = [w for w in words if w not in stopword] # <-----------------關鍵技術
words_stat = pd.Series(words).value_counts()[:20]
plt.figure(figsize=(20, 4))
plt.bar(words_stat.index, words_stat.values)
plt.show()
```
