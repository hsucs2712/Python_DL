```python
# tfidf : term() frequency -inverse document frequency
# 檢視該單字 在文章中出現次數(tf)與跨越文章出現次數(idf)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.metrics.pairwise import cosine_similarity

corpus = [
    'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
    'Is this the first document?',
]
vectorizer = CountVectorizer() # 顯示單字出現的頻率
X = vectorizer.fit_transform(corpus)


print('vocabuary', '\n', vectorizer.get_feature_names_out())

print('Sentence to vector', '\n', X.toarray() )  
```

```python
import numpy as np
# 根據 X.toarray() 對應出字典單字
words = np.array(vectorizer.get_feature_names_out())
for i in range(len(corpus)):
    print( words[(X.toarray()[i]).astype('bool')] )
```

```python
words[X.toarray()[0].astype('bool')]
```

```python
X.toarray()
```

```python
# 轉換成 tf idf
transformer = TfidfTransformer()
```

```python
tfidf = transformer.fit_transform(X)
tfidf.toarray().shape # 四篇文章 所有的單字 
print(tfidf.toarray()) # 該部分類似 CountVectorizer() 顯示單字出現的頻率，加上權重
```

```python
cosine_similarity(tfidf[0], tfidf).ravel() # 找出第幾篇 文章 跟其他的相似度(包含自己)
```

```python
#     'This is the first document.',
#     'This document is the second document.',
#     'And this is the third one.',
#     'Is this the first document?',
```

```python
# 根據文章找出哪一筆文章最相似
base = 3
similarity_subject = np.array(cosine_similarity(tfidf[base], tfidf)).ravel()
similarity_subject
```

```python
# 新文章步驟一
news = ['That one is my third document']
news_x = vectorizer.transform(news)
news_x.toarray()
```

```python
# 新文章步驟二
news_x_tfidf = transformer.transform( news_x ).toarray()
news_x_tfidf
```

```python
# 新文章步驟三
similarity_subject_news = np.array(cosine_similarity(news_x_tfidf, tfidf)).ravel()
similarity_subject_news
```

```python
corpus[np.argmax(similarity_subject_news)]
```
