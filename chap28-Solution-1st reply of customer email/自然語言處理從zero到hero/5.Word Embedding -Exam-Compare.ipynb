```python
glove = []
with open('glove.6B.100d.txt', encoding='utf-8') as f:
    glove = f.readlines()
```

```python
import numpy as np
glove_embedded = dict([ (line.split()[0], np.array(line.split()[1:], dtype='float32')) for line in glove])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
glove_embedded['sushi'] , glove_embedded['sashimi']
print(np.array(cosine_similarity(glove_embedded['sushi'].reshape(1,-1), glove_embedded['sashimi'].reshape(1,-1))).ravel())
print(np.array(cosine_similarity(glove_embedded['sushi'].reshape(1,-1), glove_embedded['dumping'].reshape(1,-1))).ravel())
print(np.array(cosine_similarity(glove_embedded['sushi'].reshape(1,-1), glove_embedded['sake'].reshape(1,-1))).ravel())
```

```python
words = list(glove_embedded.keys())
words[:10]
```

```python
# 利用 cosine_similarity 比對 指定單字與 所有單字的 空間距離
compare = np.array(cosine_similarity(glove_embedded['sushi'].reshape(1,-1), list(glove_embedded.values())  )).ravel()
```
