```python
import re
s = "Tim's phone numbers are 12345-41521 and 78963-85214"
match = re.findall(r'\d', s)
if match:
    print(match)
```

```python
import re
s = "Tim's phone numbers are 12345-41521 and 78963-85214"
match = re.findall(r'\d{5}', s)
if match:
    print(match)
```

```python
import re
s = "Tim's phone numbers are 12345-41521 and 78963-85214"
match = re.findall(r'\s\w+', s) # \s 空白 \w+ 英數\w多個+
if match:
    print(match)
```

```python
import re
pattern = re.compile(r'\d+')  
result1 = pattern.findall('hello 123456 789')
result2 = pattern.findall('one1two2three3four4', 0, 10)
print(result1)
print(result2)
```

```python
import re
str1="abcd abcdd abcb abc"
print(re.findall(r"abcd?", str1))# 可以有一個d 或是不用有 d
print(re.findall(r"abcd*",str1)) # 可以有一個d 或是不用有 d 或是多個 d
print(re.findall(r"abcd+",str1)) # 需要有d 出現一次到多次
```

```python
print(re.findall(r"[a-c]{3}", str1))  # abc組合僅有三個字
print(re.findall(r"[a-c]{1,4}", str1))# abc組合可以從一個字到四個字
print(re.findall(r"[a-c]{4,}",str1))  # abc組合至少四個字(包含)
```
